### Click The Route Link one by one extracted the Bus name, Departing time , Ending time, Duration, Bus Type, Rating , Price and Seat Availability

```python
# Initialize dictionary for storing bus details
bus_list = []

# Initialize the WebDriver
driver = webdriver.Chrome()

def scroll(driver):
    """Scrolls to the bottom of the page to load dynamic content."""
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

def extract_time(name, url):
    """Extracts bus timing and other details from the provided URL."""
    try:
        driver.get(url)
        driver.maximize_window()
        time.sleep(4)
        buttons = driver.find_elements(By.CLASS_NAME, 'p-left-10')
        if buttons and len(buttons) > 1:
            for button in reversed(buttons):
                button.click()
                time.sleep(2)
        scroll(driver)
        time.sleep(4)
        select = '#result-section > div > div > ul > div, #result-section > div > ul > div'
        # Wait for the last set of elements to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, select))
        )
        # Locate all bus items
        bus_items = driver.find_elements(By.CSS_SELECTOR, select)
        # Iterate through each bus item and extract details
        for bus in bus_items:
            try:
                bus_name = bus.find_element(By.CSS_SELECTOR, 'div.lh-24.d-color').text
                bus_type = bus.find_element(By.CSS_SELECTOR, 'div.bus-type.f-12.m-top-16.l-color.evBus').text
                duration = bus.find_element(By.CSS_SELECTOR, 'div.dur.l-color.lh-24').text
                departing_time = bus.find_element(By.CSS_SELECTOR, 'div.f-19.d-color.f-bold').text
                reaching_time = bus.find_element(By.CSS_SELECTOR, 'div.disp-Inline').text
                try:
                    star_rating = bus.find_element(By.CSS_SELECTOR, 'div.rating span').text
                except:
                    star_rating = None
                price = bus.find_element(By.CSS_SELECTOR, 'div.fare').text
                seat_availability = bus.find_element(By.CSS_SELECTOR, 'div.seat-left').text
                bus_list.append({
                    "Route Name": name,
                    "Route Link": url,
                    "Bus Name": bus_name,
                    "Bus Type": bus_type,
                    "Departing Time": departing_time,
                    "Reaching Time": reaching_time,
                    "Duration": duration,
                    "Star Rating": star_rating,
                    "Price": price,
                    "Seat Availability": seat_availability
                })
            except Exception as e:
                # print(f"Error extracting details for a bus: {e}")
                continue
    except Exception as e:
        print(f"Error processing route {name}: {e}")
        pass

# List of route links (assuming route_data is already defined and populated)
for i, link in enumerate(route_data['Urls']):
    route_name = route_data['Route Name'][i]
    print(f'Processing route "{route_name}" ({i + 1}/{len(route_data["Urls"])}): {link}')
    extract_time(route_name, link)

# Close the WebDriver
driver.quit()

# Print the extracted bus data
for bus in bus_list:
    print(bus)

# Convert the route data to a DataFrame and save it as a CSV file
bus_data = pd.DataFrame(bus_list)
bus_data.to_csv('bus_data.csv', index=False)
print('Successful to download Route data')
```

### Replace the correct order dataset

```python
df['Price'] = df['Price'].str.replace('INR ', '')

df['Star Rating'] = df['Star Rating'].replace('*','0')

df.head(3)
```

```python

df['Departing Time'] = pd.to_datetime(df['Departing Time'], format='%I:%M %p').dt.strftime('%I:%M %p')
df['Reaching Time'] = pd.to_datetime(df['Reaching Time'], format='%I:%M %p').dt.strftime('%I:%M %p')
# df['Departing Time'] = pd.to_datetime(df['Departing Time'], format='%H:%M').dt.strftime('%I:%M %p')
# df['Reaching Time'] = pd.to_datetime(df['Reaching Time'], format='%H:%M').dt.strftime('%I:%M %p')

df.head(3)
```

### Identify  the Bus Type


```python
# Define a function to classify the bus type with handling for complex types
def classify_bus_type(bus_type):
    if isinstance(bus_type, str):
        bus_type = bus_type.lower()  # Convert to lower case for uniform comparison
        
        
        # Determine A/C or Non-A/C
        if 'a/c' in bus_type:
            ac_type = 'A/C'
        elif 'non a/c' in bus_type or 'non-a/c' in bus_type:
            ac_type = 'Non-A/C'
        else:
            ac_type = 'Unknown A/C'
        
        # Determine Seater, Sleeper, Semi Sleeper, or both
        if 'semi sleeper' in bus_type:
            seat_type = 'Semi Sleeper'
        elif 'sleeper' in bus_type and 'seater' in bus_type:
            seat_type = 'Seater/Sleeper'
        elif 'sleeper' in bus_type:
            seat_type = 'Sleeper'
        elif 'seater' in bus_type:
            seat_type = 'Seater'
        else:
            seat_type = 'Unknown Type'
        
        # Combine results
        return f'{ac_type} {seat_type}'
    
    return 'Unknown'  # For NaN or unexpected values

# Apply the function to the 'Bus Type' column
df['Bus Type (Categorized)'] = df['Bus Type'].apply(classify_bus_type)

```

```python
df = df.drop(columns=['Route Link'])

df.head(3)
```

```python
df['Seat Availability'] = df['Seat Availability'].str.replace(' Seats available', '', regex=False)
df['Seat Availability'] = df['Seat Availability'].str.replace(' Seat available', '', regex=False)
df.head(3)
```

### Splite The Route Name

```python
df[['Starting Point', 'Ending Point']] = df['Route Name'].str.split(' to ', expand=True)
df.head(3)
```

```python
# Reorder the columns: Moving "Starting Point" and "Ending Point" to the first positions
new_column_order = ['Starting Point', 'Ending Point'] + [col for col in df.columns if col not in ['Starting Point', 'Ending Point']]
df = df[new_column_order]
df.head(3)
```
